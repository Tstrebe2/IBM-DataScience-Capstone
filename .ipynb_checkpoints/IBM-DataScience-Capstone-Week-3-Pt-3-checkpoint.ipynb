## Part 1: Data Preparation
#### I've decided to use the Pandas read_html function as it is simple and elegant.

```python
can_postal_link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
neighborhoods = pd.read_html(can_postal_link)[0]
# Use boolean mask to remove unassigned Borough's
neighborhoods = neighborhoods[neighborhoods.Borough != 'Not assigned']
# Group dataframe and aggregate neighborhood into comma-separated list
neighborhoods = (neighborhoods.groupby(['Postal Code', 'Borough'])
    .apply(lambda grp: ', '.join(grp.Neighbourhood.tolist()))
    .to_frame()
    .rename({0:'Neighbourhood'}, axis=1)
    .reset_index())
# Replace unassigned neighborhoods with borough names
neighborhoods['Neighbourhood'] = neighborhoods.apply(lambda x: x.Borough 
                                                     if x.Neighbourhood == 'Not assigned' 
                                                     else x.Neighbourhood, axis=1)

neighborhoods.head()
```

#### Next we will obtain the lat-long of each neighborhood postal code for our analysis.

```python
def get_lat_long(postal_code): 
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
      lat_lng_coords = g.latlng

    latitude = lat_lng_coords[0]
    longitude = lat_lng_coords[1]

    return latitude, longitude

lat_lng = neighborhoods['Postal Code'].apply(get_lat_long)

neighborhoods['Latitude'] = lat_lng.apply(lambda x: x[0])
neighborhoods['Longitude'] = lat_lng.apply(lambda x: x[1])
```

```python
pd.set_option('display.max_rows', 500)

neighborhoods.sort_values(['Borough', 'Postal Code'])
```

#### Let visualize our data by creating a folium map!

```python
def map_all(neighborhoods):
    map_all = folium.Map(location=[neighborhoods['Latitude'].median(), neighborhoods['Longitude'].median()], zoom_start=10)

    # add markers to map
    for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], 
                                               neighborhoods['Longitude'], 
                                               neighborhoods['Borough'], 
                                               neighborhoods['Neighbourhood']):
        label = '{}, {}'.format(neighborhood, borough)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color='blue',
            fill=True,
            fill_color='#3186cc',
            fill_opacity=0.7,
            parse_html=False).add_to(map_all)  

    return map_all

map_all(neighborhoods)
```

#### I want to obtain all venues and respective categories within a 500 meter radius of each neighborhood.

```python
def explore_neighborhoods(neighborhoods, radius=500):
    venues_list = []
    for neighborhood, latitude, longitude in zip(neighborhoods['Neighbourhood'], neighborhoods['Latitude'], neighborhoods['Longitude']):
        url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, radius, LIMIT)
        
        try:
            results = requests.get(url).json()
            
            if 'response' in results.keys():
                if 'groups' in results['response'].keys():
                    if len(results['response']['groups']) > 0:
                        if 'items' in results['response']['groups'][0]:
                            results = results['response']['groups'][0]['items']

                            venues_list.append([(
                                neighborhood,
                                latitude,
                                longitude,
                                v['venue']['name'],
                                v['venue']['location']['lat'], 
                                v['venue']['location']['lng'],  
                                v['venue']['categories'][0]['name']) for v in results])
        except:
            pass
                        
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return nearby_venues
    
nearby_venues = explore_neighborhoods(neighborhoods)
```

#### 181 categories translates to 181 potential features for our K-Means algorithm. Lets turn categories these into numeric representations by using one-hot-encoding.

```python
torronto_one_hot = pd.get_dummies(nearby_venues['Venue Category']).join(nearby_venues['Neighborhood'], lsuffix='oh_')

fixed_columns = [torronto_one_hot.columns[-1]] + list(torronto_one_hot.columns[:-1])

torronto_one_hot = torronto_one_hot[fixed_columns]

torronto_one_hot.head()
```

```python
torronto_one_hot.shape
```

#### Lets take our one-hot data and generate frequrencies of venue categories for each neighborhood.

```python
torronto_grouped = torronto_one_hot.groupby('Neighborhood').mean().reset_index()
torronto_grouped
```

## Part 2: Data Analysis
#### Before we clusture our dataset, I want to see differences between our neighborhoods by looking at the top 5 venue frequencies for each.

```python
num_top_venues = 5

for hood in torronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = torronto_grouped[torronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### The top 5 was good, but I want to tabulate the top ten most common venues for each neighborhood to obtain a broader inference.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = torronto_grouped['Neighborhood']

for ind in np.arange(torronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(torronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Part 3: Clustering
#### I am going to segment my dataset into 5 clusters to see which neighborhoods are most alike and which are most different

```python
# set number of clusters
kclusters = 5

torronto_grouped_clustering = torronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(torronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
torronto_merged = (torronto_grouped.join([pd.Series(kmeans.labels_, name='Cluster Labels')])
                   .merge(neighborhoods[['Neighbourhood', 'Latitude', 'Longitude']], left_on='Neighborhood', right_on='Neighbourhood'))

torronto_merged.head()
```

#### Now that our dataset is labeled, lets look at a visual representation on a map.

```python
map_clusters = folium.Map(location=[torronto_merged.Latitude.median(), torronto_merged.Longitude.median()], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(torronto_merged['Latitude'], 
                                  torronto_merged['Longitude'], 
                                  torronto_merged['Neighborhood'], 
                                  torronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)],
        fill=True,
        fill_color=rainbow[int(cluster)],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
